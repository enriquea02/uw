### 4. Sequences via recurrence relations

Let $x_0 = 1$ and $x_{i+1} = - a x_i$ where $0 < a < 1$.

a) Plot the first 20 points in $(x_n)$ when $a=\frac{1}{2}$.

b) Show that $(x_n)$ converges to 0.

```python
# Set constant "a"
Fn4_const = (1/2)
# Initialize sequence array
Y4 = np.array([sm.Integer(n) for n in range(0,20)])
# Initialize first term in sequence to x_0 = 1
Y4[0] = 1
# implement recursive formula
Fn4 = lambda n : -(Fn4_const*n)
# solve recursive formula
for n in range(0,19):
    Y4[(n+1)] = Fn4(Y4[n])

# Plot recursive sequence
plt.scatter([sm.Integer(n) for n in range(0,20)],Y4, c="yellow",
    marker ="o", 
    edgecolor ="black", 
    s = 50)
plt.legend(["x_{i+1} = -ax_i"])
plt.xlabel("i")
plt.ylabel("x_i")
plt.title("Recursive Function Graph\n");
plt.show
```

### 8. Definition of Derivative

Use the definition of the derivative to find the derivatives of the following functions.

a) $f(x) = x^3$, for any $x$

b) $g(x) = 1/x$, for $x \neq 0$

c) $h(x) = \sqrt{x}$, for $x > 0$

d) $k(x) = x^{-1/2}$ for $x > 0$.

a) $$f'(x) = \lim\limits_{x \rightarrow c} \frac{x^3 - c^3}{x - c}$$
Factor numerator, $$x^3 - c^3 = (x-c)(x^2 + cx + c^2)$$
Cross out like terms and simplify, $$f'(x) = \lim\limits_{x \rightarrow c} (x^2 + cx + c^2) = (c^2 + c^2 + c^2) = 3c^2$$
<br><br>
b) $$g'(x) = \lim\limits_{x \rightarrow c} \frac{\frac{1}{x}-\frac{1}{c}}{x-c}$$
Set fractions in numerator to a common denominator, $$g'(x) = \frac{\frac{-(x-c)}{xc}}{x-c}$$
Cross out like terms and simplify, $$g'(x) = - \frac{1}{xc} = - \frac{1}{c^2}$$
<br><br>
c) $$h'(x) = \lim\limits_{x \rightarrow c} \frac{\sqrt{x}-\sqrt{c}}{x-c}$$
Multiply numerator and denominator by $\sqrt{x}+\sqrt{c}$, $$h'(x) = \frac{(x-c)+(\sqrt{xc}-\sqrt{xc})}{(x-c)(\sqrt{x}+\sqrt{c})}$$
Cross out like terms and simplify, $$h'(x) = \frac{1}{\sqrt{x}+\sqrt{c}} = \frac{1}{2\sqrt{c}}$$
<br><br>
d) $$k'(x) = \lim\limits_{x \rightarrow c} \frac{x^{-1/2}-c^{-1/2}}{x-c}$$
Multiply numerator and denominator by $x^(-1/2) + c^(-1/2)$ $$k'(x) = \frac{x^{-1}-xc^{-1/2}+xc^{-1/2}-c^{-1}}{(x-c)(x^{-1/2}+c^{-1/2})}$$
Cross out like terms and simplify, $$k'(x) = \frac{-1}{(xc)(x^{-1/2}+c^{-1/2})}$$
Pull out factors, rearrange, and simplify $$k'(x) = \frac{-1}{\sqrt{x}\sqrt{c}(\sqrt{c}+\sqrt{x})} = \frac{-1}{2c^{3/2}}$$

### 9. The Chain Rule

Use the chain rule to find these deriviatives. Verify your results with `sympy`.

a) $f(x) = \frac{1}{1+x^2}$

b) $h(x) = (\sin x^k)^m$ for $m,k \in \mathbb{N}$

a) $$f(x) = \frac{1}{1+x^2}$$
Use the chain rule to solve. Set $f(x) = u = 1+x^2$, $$(g \circ f)'(x) = g'(f(x)) \cdot f'(x) = \frac{d}{du} \frac{1}{u} \cdot \frac{du}{dx}$$
Solve for g'(f(x)), $$g'(f(x)) = \frac{d}{du} \frac{1}{u} = -\frac{1}{u^2}$$
Substitute $u$, $$g'(f(x)) = -\frac{1}{u^2} = -\frac{1}{(1+x^2)^2}$$
Solve for f'(x), $$f'(x) = \frac{du}{dx} = 2x$$
Multiply g'(f(x)) $\cdot$ f'(x), $$(g \circ f)'(x) = g'(f(x)) \cdot f'(x) = -\frac{1}{(1+x^2)^2} \cdot 2x = -\frac{2x}{(1+x^2)^2}$$ 
<br><br>
b) $$h(x) = (\sin x^k)^m$$
Use the chain rule to solve. Set $h(x) = u = \sin x^k$, $$(g \circ h)'(x) = g'(h(x)) \cdot h'(x) = \frac{d}{du} \frac{1}{u} \cdot \frac{du}{dx}$$
Solve for g'(f(x)), $$g'(h(x)) = \frac{d}{du} u^m = mu^{m-1}$$
Substitute $u$, $$g'(h(x)) = mu^{k-1} = m(\sin x^k)^{m-1}$$
Solve for h'(x), $$h'(x) = sin x^k = kx^{k-1}cos(x^k)$$
Multiply g'(h(x)) $\cdot$ h'(x), $$(g \circ h)'(x) = g'(h(x)) \cdot h'(x) = m(\sin x^k)^{m-1} \cdot kx^{k-1}cos(x^k) = mkx^{k-1}(\sin x^k)^{m-1}cos(x^k)$$
