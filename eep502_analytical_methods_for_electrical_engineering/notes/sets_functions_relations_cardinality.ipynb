## Notes
### Sets
Proofs for sets can be written using ZFC axioms written by Zermelo and Fraenkel <br>
<br>
All sets include the $ \varnothing $ subset <br>
<br>
$ \mathbb{N} $ in set theory inclues zero, so ignore $ \mathbb{N}_0 $
$ \mathbb{Z} $ comes from "zahl" (German for number). $ \mathbb{Z} $ includes zero. <br>
$ \mathbb{Q} $ comes from quotient <br>
$$ \mathbb{Q} = p/q : p,q \isin \mathbb{Z} \text{ and } q \neq 0 $$
<br>
Complex analysis will not be covered in real analysis. For fun, set builder notation for entire set of complex numbers: <br>

$$ \mathbb{C} = x + iy : \text{ x,y } \isin \R $$ 

$$ \mathbb{Z}_+ != \mathbb{Z}_{++}: \mathbb{Z}_+ \text{ includes non-negative integer range and zero, } \mathbb{Z}_{++} \text{ includes non-negative intergers but not zero } $$
$$ \mathbb{Z}_{--} == \mathbb{Z}_- : \text{ are same, because zero is not included in negative integer range } $$
