## Notes
### Sequences
Sequence $x_n$ is an ordered list of terms. <br>
Sequence $\neq$ set. Unlike sets, terms can repeat. <br>
Sequence of real numbers is a function. So plot it bitch! $$ \text{sequence } x_n \text{ is a } f:\mathbb{N} \rightarrow \mathbb{R} \text{, where } x_n = f(n) $$ <br>
Sequence out can be any object, but focus on $x_n \in \mathbb{R}$ <br>
